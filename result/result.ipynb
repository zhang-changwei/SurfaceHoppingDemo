1. Simple Avoid crossing model (SAC)

the adabatic potential suface and the nonadabatic coupling look like this.  


```python
import numpy as np
from matplotlib import pyplot as plt

qd = np.genfromtxt('qdsac.txt')
sh = np.genfromtxt('shsac.txt')

k = qd[:,0]
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True)
ax1.plot(k, qd[:,3], ls='-', marker='.', label='QD')
ax1.plot(k, sh[:,3], ls='-', marker='.', label='SH')
ax1.set_ylabel('Lower Trans.')
ax2.plot(k, qd[:,4], ls='-', marker='.', label='QD')
ax2.plot(k, sh[:,4], ls='-', marker='.', label='SH')
ax2.set_ylabel('Upper Trans.')
ax3.plot(k, qd[:,1], ls='-', marker='.', label='QD')
ax3.plot(k, sh[:,1], ls='-', marker='.', label='SH')
ax3.set_ylabel('Lower Reflect.')
ax3.set_xlabel('k (a.u.)')
ax1.legend()
plt.show()
```

3. Extended Coupling Reflection (ECR)   


```python
import numpy as np
from matplotlib import pyplot as plt

qd = np.genfromtxt('qdecr.txt')
sh = np.genfromtxt('shecr.txt')

k = qd[:,0]
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, sharex=True)
ax1.plot(k, qd[:,3], ls='-', marker='.', label='QD')
ax1.plot(k, sh[:,3], ls='-', marker='.', label='SH')
ax1.set_ylabel('Lower Trans.')
ax2.plot(k, qd[:,4], ls='-', marker='.', label='QD')
ax2.plot(k, sh[:,4], ls='-', marker='.', label='SH')
ax2.set_ylabel('Upper Trans.')
ax3.plot(k, qd[:,1], ls='-', marker='.', label='QD')
ax3.plot(k, sh[:,1], ls='-', marker='.', label='SH')
ax3.set_ylabel('Lower Refl.')
ax4.plot(k, qd[:,2], ls='-', marker='.', label='QD')
ax4.plot(k, sh[:,2], ls='-', marker='.', label='SH')
ax4.set_ylabel('Upper Refl.')
ax4.set_xlabel('k (a.u.)')
ax1.legend()
plt.show()
```

Tips:  
For QD simulation, many parameters have to be toned to get a reasonable result.
For example, the convergence condition and the time step have to be toned to trade off accuracy for speed. And the number of grid in real space has also to be large enough to get a reasonable result in high initial kinetic energy case.

Details:
1. For SAC, DAC, x0 = -5.0 a.u. for ECR, x0 = -10.0 a.u.
2. For SH calculation, ntraj = 2000 and dt = 0.5 a.u.
3. For QD calculation, ndvr = 4096 and 2L = 200 a.u.
4. oneshot method is applied for SAC and DAC model at t = 10^5/k0, which is the time for a classical partical reach roughly L/2 point.
5. compute method is applied for ECR model given the condition that the paticle will have a large acceralation after transmits to the right side of the lower level.
